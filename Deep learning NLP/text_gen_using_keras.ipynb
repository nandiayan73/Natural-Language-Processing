## tokenize the text and create sequences with keras

```python
def read_file(filepath):
    with open(filepath) as f:
        str_text=f.read()
    return str_text
```

```python
read_file("moby_dick_four_chapters.txt")
```

```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm',disable=['parser','tagger','ner'])
# nlp = spacy.load('en',disable=['parser', 'tagger','ner'])


nlp.max_length=1198623
```

Removing extra unwanted string from text

```python
def separate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if( token.text) not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
d=read_file('moby_dick_four_chapters.txt')
tokens=separate_punc(d)
```

# 25 words -> predict the 26th word

```python
train_len=25+1
text_sequences=[]

for i in range(train_len,len(tokens)):
    seq=tokens[i-train_len:i]
    text_sequences.append(seq)
print((text_sequences[0]))
```

# Keras Tokenization

```python
from keras.preprocessing.text import Tokenizer

tokenizer=Tokenizer()
tokenizer.fit_on_texts(text_sequences)
sequences=tokenizer.texts_to_sequences(text_sequences)
```

```python
print((sequences[0]))
```

```python
print(tokenizer.index_word)#common words with index number
```

```python
print(tokenizer.word_counts)#words with their repetitions
```

```python
vocabulary_size=len(tokenizer.word_counts)#total unique words
```

changing squences into numpy array

```python
import numpy as np
```

```python
sequences=np.array(sequences)
```

```python
print(sequences)
```

## Creating our LSTM based model

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding
```

```python
print(sequences[:,:-1])
```

```python
print(sequences[:,-1])
```

```python
x=sequences[:,:-1]
print(x)
```

```python
from keras.utils import to_categorical
y=sequences[:,-1]
# One hunt encoding
y=to_categorical(y,num_classes=vocabulary_size+1)
print(y)
```

```python
print(sequences[:,-1])
print(y.shape)
```

```python
seq_len=x.shape[1]
print(vocabulary_size)
```

```python
def create_model(vocabulary_size,seq_len):
    model=Sequential()
    model.add(Embedding(vocabulary_size,15,input_length=seq_len))
    model.add(LSTM(150,return_sequences=True))
    model.add(LSTM(150))
    model.add(Dense(150,activation='relu'))
    model.add(Dense(vocabulary_size,activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.summary()
    return model
```

```python
model=create_model(vocabulary_size+1,seq_len)
```

```python
from pickle import dump,load
```

```python
model.fit(x,y,batch_size=128,epochs=2,verbose=1)
```

```python
model.save('result.h5')
```
